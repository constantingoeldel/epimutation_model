```python

import matplotlib.pyplot as plt
import numpy as np
import os


# Get list of all files in the log directory
log_files = os.listdir("log")

# Create a figure
plt.figure(figsize=(10, 5))

# For each file in the directory
for i, log_file in enumerate(log_files):
    # Read data from file
    with open(f"log/{log_file}", "r") as f:
        data = f.read().split("\n")
        # Skip header
        data = data[1:-1]
        data = [d.split(",") for d in data]
        data = data[:-1] if data[-1] == '' else data  # Remove last empty line if it exists

    # Convert to numpy array
    data = np.array(data, dtype=np.float32)

    # Plot the training loss, test loss, and time against the epoch in the same plot
    plt.plot(data[:, 0], data[:, 2], label=f"{log_file.replace('.txt', '')} - Training loss")
    plt.plot(data[:, 0], data[:, 3], label=f"{log_file.replace('.txt', '')} - Test loss")

# Set labels, legend and show the plot
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.yscale("log")
plt.legend()
plt.tight_layout()
plt.show()
```
