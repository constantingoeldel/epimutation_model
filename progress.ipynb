```python

import matplotlib.pyplot as plt
import numpy as np
import os
import polars as pl


# Get list of all files in the log directory
log_files = os.listdir("log")

# Create a figure
plt.figure(figsize=(10, 5))

# For each file in the directory
for i, log_file in enumerate(log_files):
    df = pl.read_csv(f"log/{log_file}", separator=",")
    # Convert to numpy array
    data = np.array(df, dtype=np.float32)

    # Plot the training loss, test loss, and time against the epoch in the same plot
    plt.plot(data[:, 1], data[:, 2], label=f"{log_file.replace('.txt', '')} - Training loss")
    plt.plot(data[:, 1], data[:, 3], label=f"{log_file.replace('.txt', '')} - Test loss")

# Set labels, legend and show the plot
plt.xlabel("Step")
plt.ylabel("Loss")
plt.yscale("log")
plt.legend()
plt.tight_layout()
```
